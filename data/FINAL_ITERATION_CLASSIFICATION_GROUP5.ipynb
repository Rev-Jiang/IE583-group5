We have now clustered both datasets and we will now check the validity of the clustering by using a classification algorithm to classify articles by their respective clusters.

```python
#import libraries
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix

seed = 1234
```

## tf-idf classification

```python
tfidf = pd.read_csv('tfidf_clustered.csv')
tfidf.head()
```

```python
preds = tfidf['preds']
tfidf = tfidf.iloc[:,:-1]
tfidf
```

```python
# Break up into independent training and test sets for evaluation purposes
# I get super mixed up with X and y so I use input/output with their respective independent set
input_train, input_test, output_train, output_test = train_test_split(tfidf, preds, test_size=0.3, random_state = seed)
```

```python
randomForest_tfidf = RandomForestClassifier(n_estimators=100).fit(input_train, output_train)
```

```python
tfidf_preds_test = randomForest_tfidf.predict(input_test)
```

```python
#confusion matrix to assess random forest model
conMat_tfidf = confusion_matrix(output_test, tfidf_preds_test)
conMat_tfidf
```

```python
# Accuracy of the classification model
accuracy_tfidf = accuracy_score(output_test, tfidf_preds_test)
accuracy_tfidf
```

So as you can see the classifier worked incredibly well, but it is not incredibly impressive since 88% of the points belonged to a single cluster, yet there were only two misclassifications, impressive!

## document-term classification

```python
title = pd.read_csv('title_clustered.csv')
#eliminate index
title = title.iloc[:, 1:]
title
```

```python
title_classification = title.iloc[:, 4:]
preds_title = title_classification['preds']
title_classification = title_classification.iloc[:,:-1]
title_classification
```

```python
train_in, test_in, train_out, test_out = train_test_split(title_classification, preds_title, test_size=0.3, random_state = seed)
```

```python
randomForest_title = RandomForestClassifier(n_estimators=100).fit(train_in, train_out)
```

```python
title_preds_test = randomForest_title.predict(test_in)
```

```python
#confusion matrix to assess random forest model
conMat_title = confusion_matrix(test_out, title_preds_test)
conMat_title
```

```python
# Accuracy of the classification model
accuracy_title = accuracy_score(test_out, title_preds_test)
accuracy_title
```
